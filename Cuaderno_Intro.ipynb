## Precio de un traje

### Diagrama e flujo


image.png

### Pseudocodigo


```
Algoritmo precioTraje
	Leer precio
	total = 0
	Si precio>2500 Entonces
		descuento = 0.15
		rebaja <- descuento*precio
		total = precio-rebaja
		Escribir total
	SiNo
		descuento = 0.08
		rebaja <- descuento*precio
		total = precio-rebaja
		Escribir total
	FinSi
FinAlgoritmo

```



### Algoritmo

````python
def precioTraje(precio):
  total=0
  if precio>2500:
    descuento=0.15
    rebaja=descuento*precio
    total=precio-rebaja
    return total
  else:
    descuento=0.08
    rebaja=descuento*precio
    total=precio-rebaja
    return total
````

### Pseudocodigo



```
Algoritmo presupuestoEvento
	Leer numPersonas
	precioPersona <- 0
	Si numPersonas>5 & numPersonas<=10 Entonces
		precioPersonas = 8500
	SiNo
		Si numpersonas>11 Entonces
			precioPersona = 7500

	FinSi
	total <- numPersonas*precioPersona
	Escribir total

```



### Algoritmo 

````python
def presupuestoEvento(numPersonas):
  total=0
  if numPersonas>5 and numPersonas>= 10:
    precioPersonas=8500
    total=precioPersonas*numPersonas
    return total
  elif numPersonas>11:
    precioPersonas=7500
    total=precioPersonas*numPersonas
    return total
  else:
    precioPersonas=9500
    total=precioPersonas*numPersonas
    return total

````
