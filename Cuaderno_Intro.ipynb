## Factorial

### Diagrama de flujo

image.png

### Pseudocodigo



```
Algoritmo factorial
	Leer num
	Si num>0 Entonces
		digito <- 1
		total <- 1
		Mientras num>=digito Hacer
			total <- total * digito
			digito <- digito+1
		FinMientras
		Escribir total
	SiNo
		Escribir Error
	FinSi

```




### Algoritmo


````python
def factorial(num):
  if num>0:   
    digito=1
    total=1
    while num>=digito:
      total=total*digito
      digito+=1
    return total
  else:
    return "Error"
````
